# Uber Analytics Data Cleaning

This notebook provides a beginner-friendly workflow for cleaning data using `pandas`. The steps include loading the dataset, inspecting it, handling missing values, removing duplicates, converting data types, and standardizing column names.


## 1. Import Required Libraries

Import `pandas` and any other necessary libraries for data cleaning.


## 2. Load the Dataset

Read the dataset into a `pandas` DataFrame. If the file is compressed, `pandas` can read it directly.


```python
data_path = 'data/raw/uber_raw_data.csv.zip'

try:
    df = pd.read_csv(data_path)
    print(f'Data loaded successfully from: {data_path}')
except FileNotFoundError:
    print(f'File not found: {data_path}')
    df = pd.DataFrame()
```

## 3. Inspect the Data

View the first rows, check the shape, and use `info()` and `describe()` to understand the dataset.


```python
df.head()
```

```python
print('Shape:', df.shape)
print('\nInfo:')
df.info()

print('\nSummary statistics:')
df.describe(include='all')
```

## 4. Handle Missing Values

Identify missing values and demonstrate how to fill or drop them using pandas.


```python
missing_counts = df.isna().sum().sort_values(ascending=False)
print('Missing values by column:')
print(missing_counts[missing_counts > 0])

# Example strategies:
# - Drop rows where a critical column is missing
# - Fill missing values with a default or derived value

# Example: drop rows where all values are missing
cleaned_df = df.dropna(how='all')
print('\nRows after dropping fully empty rows:', cleaned_df.shape)
```

## 5. Remove Duplicate Rows

Check for and remove duplicate rows from the DataFrame.


```python
duplicate_count = cleaned_df.duplicated().sum()
print('Duplicate rows found:', duplicate_count)

cleaned_df = cleaned_df.drop_duplicates()
print('Shape after removing duplicates:', cleaned_df.shape)
```

## 6. Convert Data Types

Convert columns to appropriate data types using `astype()` or `to_datetime()`.


```python
possible_date_cols = [col for col in cleaned_df.columns if 'date' in col.lower() or 'time' in col.lower()]
print('Possible date/time columns:', possible_date_cols)

if possible_date_cols:
    for col in possible_date_cols:
        try:
            cleaned_df[col] = pd.to_datetime(cleaned_df[col], errors='coerce')
            print(f'Converted {col} to datetime.')
        except Exception as e:
            print(f'Could not convert {col}:', e)
```

## 7. Standardize Column Names

Rename columns to have consistent formatting for easier use later.


```python
cleaned_df.columns = (
    cleaned_df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)

print('Standardized column names:')
print(cleaned_df.columns.tolist())
```

## Save Cleaned Data (Optional)

Save the cleaned dataset for later use.


```python
output_path = 'data/uber_cleaned_data.csv'
cleaned_df.to_csv(output_path, index=False)
print(f'Cleaned data saved to: {output_path}')
```
